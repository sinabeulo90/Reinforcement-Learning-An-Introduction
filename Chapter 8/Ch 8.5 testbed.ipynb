```python
import numpy as np
import matplotlib.pyplot as plt

from tqdm import trange

np.random.seed(42)
random_state = np.random.get_state()
```

```python
# for figure 8.7, run a simulation of 2 * @b steps
def b_steps(b):
    # set the value of the next b states
    # it it not clear how to set this
    distribution = np.random.rand(b)

    # true value of the current state
    true_v = np.mean(distribution)

    sample = []
    errors = []

    # sample 2b steps
    for t in range(2 * b):
        v = np.random.choice(distribution)
        sample.append(v)
        errors.append(np.abs(np.mean(sample) - true_v))
    
    return errors
```

```python
def figure_8_7():
    runs = 100
    branch = [2, 10, 100, 1000]
    for b in branch:
        errors = np.zeros((runs, 2 * b))
        for run in trange(runs):
            errors[run] = b_steps(b)
        errors = errors.mean(axis=0)
        x_axis = (np.arange(len(errors)) + 1) / float(b)
        plt.plot(x_axis, errors, label="b = {:d}".format(b))
    
    plt.xlabel("number of computations")
    plt.xticks([0, 1.0, 2.0], ["0", "b", "2b"])
    plt.ylabel("RMS error")
    plt.legend()
    plt.show()
```

```python
figure_8_7()
```
